```python
import os
import random
import math
import hangman_words

def generate_quiz_word(choosen_word, guess_word):
    for num in range(math.ceil(len(choosen_word)/2)):
        index = random.randint(1, len(choosen_word) - 1)
        guess_word[index] = '_'
    return guess_word
```

```python
def guess_game(choosen_word, guess_word, lives, stages):
    while '_' in guess_word and lives>=0:
        letter = input("Enter the letter here : ").casefold()
        os.system('cls')
        right_guess = False
        
        if letter in choosen_word:
            for index in range(len(choosen_word)):
                if guess_word[index] == '_':
                    if letter == choosen_word[index]:
                        guess_word[index] = letter
                        right_guess = True
            if right_guess == False:
                print(f"\"You choose \'{letter}\' which is already present in the word, you lose a live.\"")
                print(stages[lives])
                print(f"Lives : {lives}")
                lives -= 1
                
        else:
            print(f"\"You choose \'{letter}\' which is not present in the word, you lose a live.\"")
            print(stages[lives])
            print(f"Lives : {lives}")
            lives -= 1
            
        print(f"\n{' '.join(guess_word)}")
        
    return guess_word, lives
```

```python
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
lives = 6

print("Welcome to hangman game.\n     Good luck\n".title())
choosen_word = random.choice(hangman_words.words_list)
guess_word = list(choosen_word)

guess_word = generate_quiz_word(choosen_word, guess_word)

print(' '.join(guess_word))

guess_word, lives = guess_game(choosen_word, guess_word, lives, stages)

    
if lives < 0:
    print("\n\"Game Over!\"")
else:
    print("\n\"You Win!\"\n")
```
